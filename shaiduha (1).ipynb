*Descriptive Statistics**: Calculate basic statistics mean, median, mode, minimum, and maximum salary, determine the range of salaries, and find the standard deviation.


```python


import pandas as pd
import numpy as np


file_path = r'/Salaries (1).csv'
data = pd.read_csv(file_path)
df_data=pd.DataFrame(data)
df_data_new = df_data[df_data['TotalPay'] >=0]

mean_val=df_data_new['TotalPay'].mean()
median_val = df_data_new['TotalPay'].median()
mode_val =df_data_new['TotalPay'].mode()
max_val = df_data_new['TotalPay'].max()
min_val = df_data_new['TotalPay'].min()
range_val=max_val-min_val
std_val= df_data_new['TotalPay'].std()
#or can use ptp(peek to peek ) function

print("the mean value is : ",mean_val)
print("the median value is :",(median_val))
print("the mode value is :",mode_val)
print("the max value is :",max_val)
print("the min value is :",min_val)
print("the range value is :",range_val)
print("the std value is :",std_val)


```

. **Basic Data Visualization**: Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

```python
#req 4
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(df_data_new['BasePay'], bins=10, color='skyblue', edgecolor='black')
plt.title('distribution of Salaries')
plt.xlabel('salary')
plt.ylabel('frequent')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#req4 pie charts to represent the proportion of employees in different departments
#to recognize the department i used joptitle column
department = df_data_new['JobTitle'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(department, labels=department.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title('proportion of employees in different departments')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


grouped_data = df_data_new.groupby('JobTitle')
summary_statistics_by_title = grouped_data['TotalPay'].agg(['count', 'mean', 'median', 'std', 'min', 'max'])

summary_statistics_by_title.sort_values(by='mean').plot(kind='barh', figsize=(12, 8))
plt.title('Summary Statistics by Job Title')
plt.xlabel('Values')
plt.ylabel('Job Title')
plt.show()
```

. **Simple Correlation Analysis**: Identify any correlation between salary and another numerical column, and plot a scatter plot to visualize the relationship.



```python
#req 6
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df_data_new[['TotalPay', 'BasePay']].corr()

print(correlation_matrix)

# Step 4: Visualize the Correlation (optional)
sns.scatterplot(x='BasePay', y='TotalPay', data=df_data_new)
plt.title('Correlation between TotalPay and BasePay')
plt.xlabel('BasePay')
plt.ylabel('TotalPay (Salary)')
plt.show()
```
